# Feedback Effect Analysis

**Research Question:** RQ1: Does LLM-generated feedback improve skill development more than score-only feedback?

## Setup

```python
import sys
sys.path.insert(0, "../platform/orchestrator")
sys.path.insert(0, "../tools/analysis")
import statistical_tests as st
```

## Wilcoxon Signed-Rank (within treatment group)

```python
for dim in st.DIMENSIONS:
    r = st.wilcoxon_within("treatment", dim)
    print(f"{dim}: p={r.get(chr(112)_value,chr(45))}, r={r.get(chr(101)ffect_size_r,chr(45))}")
```

## Mann-Whitney U (between groups at post-test)

```python
results = st.run_all()
import pandas as pd
rows = []
for dim, r in results.items():
    mw = r["mann_whitney"]
    rows.append({"dimension": r["label"], "p_adj": mw.get("p_adjusted","n/a"), "sig": mw.get("significant_adjusted",False)})
pd.DataFrame(rows)
```

## Export LaTeX Table

```python
st.export_latex(results, "results/stats_table.tex")
st.export_json(results, "results/stats.json")
print("Done")
```
